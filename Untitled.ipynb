```python
pp = PrettyPrinter()

api_key = "DpmLXIe2jzwffE2vxsZcEp8ye0rwpiddSJrJvonu"
start_date = "2020-01-22"
end_date = "2020-01-24"

class Downloader():

    def fetch_asteroids(self, api_key, start_date, end_date):
        url = "https://api.nasa.gov/neo/rest/v1/feed"
        params = {
            "api_key":api_key,
            "start_date":start_date,
            "end_date":end_date,
            "detailed":"true"
        }
        response = requests.get(url,params=params).json()
        # pp.pprint(response)
        return response


dl = Downloader()
data = dl.fetch_asteroids(api_key, start_date, end_date)
```

```python
neo = data["near_earth_objects"]
```

```python
dates = []

for date in neo:
    dates.append(date)
```

```python
dates
```

```python
asteroids_temp = []

for date in dates:
    
    for asteroid in neo[date]:
        asteroids_temp.append(asteroid)
```

```python
asteroids = defaultdict(dict)

for asteroid in asteroids_temp:
    asteroids[asteroid["id"]] = {"name": [], "approach_distance": [], "approach_date": [], "size": []}
    asteroids[asteroid["id"]]["name"] = asteroid["name"]
    asteroids[asteroid["id"]]["approach_distance"] = asteroid["close_approach_data"][0]["miss_distance"]["kilometers"]
    asteroids[asteroid["id"]]["approach_date"] = asteroid["close_approach_data"][0]["close_approach_date"]
    asteroids[asteroid["id"]]["size"] = (asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_max"] + asteroid["estimated_diameter"]["kilometers"]["estimated_diameter_min"]) / 2
```

```python
df = pd.DataFrame.from_dict(asteroids).T
df["approach_distance"] = df["approach_distance"].astype(float)
df
```
